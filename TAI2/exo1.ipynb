```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# color = 0
# image = cv2.imread("grayscale.png", 0)
# Charger l'image en niveaux de gris
image = cv2.imread("grayscale.png", 0)
```

```python
plt.imshow(image,cmap="gray")
plt.show()

def histogram(image):
    L=256
    hist= np.zeros(L)
    H = image.shape[0]    
    W = image.shape[1]
    for i in range(H):
        for j in range(W):
            hist[image[i,j]] += 1
    return hist   
```

```python
#desiner l'histogramme
L=256
bins=[i for i in range(L)]
plt.figure(figsize=(20,10))
plt.subplot(2,2,1)
plt.title("histogram")
plt.bar(bins,histogram(image))
```

```python
def mod_image(image,th):
    for y in range(image.shape[0]):
        for x in range(image.shape[1]):
            if image[y, x] > th:
                image[y, x] = 255
            else:
                image[y, x] = 0
    return image

    
```

```python
plt.imshow(mod_image(image, 123),cmap="gray")
plt.show()
```
